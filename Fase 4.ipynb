```python
import pandas as pd
import seaborn as sns
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("spotify_top_charts_22.csv")
display(df)
```

```python
sns.scatterplot(data=df, x="energy", y="acousticness")
```

```python
Analisando diferentes colunas do dataset e realiazando teste essa foi a que mais fez sentido usarmos os k-means para analisar os grupos.
```

Como saber qual é melhor número de centroids para esse gráfico? e qual padrão foi encontrado?

```python
df_group = df[["energy", "acousticness"]]
```

```python
kmeans = KMeans(n_clusters=2, n_init=10).fit(df_group)
```

```python
kmeans.labels_
```

```python
sns.scatterplot(data=df_group, x="energy", y="acousticness", hue=kmeans.labels_)
```

```python
kmeans3 = KMeans(n_clusters=3, n_init=10).fit(df_group)
```

```python
kmeans3.labels_
```

```python
sns.scatterplot(data=df_group, x="energy", y="acousticness", hue=kmeans3.labels_)
```

```python
kmeans4 = KMeans(n_clusters=4, n_init=10).fit(df_group)
```

```python
kmeans4.labels_
```

```python
sns.scatterplot(data=df_group, x="energy", y="acousticness", hue=kmeans4.labels_)
```

```python
kmeans5 = KMeans(n_clusters=5, n_init=10).fit(df_group)
```

```python
kmeans5.labels_
```

```python
sns.scatterplot(data=df_group, x="energy", y="acousticness", hue=kmeans5.labels_)
```

```python
valores_k= []

inercias = []

for i in range (1,15):
    kmeans= KMeans (n_clusters =i,  n_init=10 , random_state=0).fit(df_group) 
    valores_k.append(i)
    inercias.append(kmeans. inertia_)
```

```python
fig , ax = plt.subplots()

ax.plot(valores_k , inercias)

plt.show
```

Vemos pelos graficos e pelo metodo do cotovelo que o melhor numero de centroids fica entre 2 e 4; Vimos que pelos grafico que quanto mais a linha energy aumenta mais a coluna acousticness diminui.
